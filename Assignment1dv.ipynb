```python
import pandas as pd
import matplotlib.pyplot as plt

# load datasets
matches = pd.read_csv("matches (1).csv")
deliveries = pd.read_csv("deliveries.csv")

# display first few rows
matches.head()
```

```python
#1Match won by maximum margin of runs
max_runs = matches.loc[matches['win_by_runs'].idxmax()]
print(max_runs[['season','team1','team2','winner','win_by_runs']])

#Insight: This shows the match where Mumbai Indians won by the highest number of run(146).
# match with the highest run margin shows the most dominant batting performance in IPL history where the winning team completely outperformed the opponent.
```

```python
#2Match won by maximum wickets
max_wickets = matches.loc[matches['win_by_wickets'].idxmax()]
print(max_wickets[['season','team1','team2','winner','win_by_wickets']])

#Insight: The KKR team won by maximum wickets which is by 10 wickets# .The match with the highest wicket margin indicates a very strong chasing performance where the winning team easily achieved the target
```

```python
#3Match won by minimum margin of runs
min_runs = matches[matches['win_by_runs'] > 0].sort_values(by='win_by_runs').head(1)
print(min_runs[['season','team1','team2','winner','win_by_runs']])

#Insight: Mumbai Indians won by minimum margin of runs which is by 1 run.
```

```python
#4Match won by minimum wickets
min_wickets = matches[matches['win_by_wickets'] > 0].sort_values(by='win_by_wickets').head(1)
print(min_wickets[['season','team1','team2','winner','win_by_wickets']])

#Insight:Sunrisers Hyderabad won the match by minimum wickets which is by 1 wicket.
```

```python
#5Matches where D/L method was and wasn't applied
dl_counts = matches['dl_applied'].value_counts()
print(dl_counts)

dl_counts.plot(kind='bar')
plt.title("D/L Method Applied vs Not Applied")
plt.show()

#Insight: D/L method was applied for 737 times
# not applied for 19 times.Most IPL matches are completed normally without interruption, while only a few matches use the Duckworth-Lewis method due to rain or weather conditions.
```

```python
#6Number of matches held in each city
city_matches = matches['city'].value_counts()

print(city_matches)

city_matches.plot(kind='bar', figsize=(10,5), color="#FBCDAF")
plt.title("Matches in Each City")
plt.xlabel("City")
plt.ylabel("Number of Matches")
plt.show()

#Insight: the maximum no. of matches  are held in Mumbai(101)
# the least in Bloemfontein(1)
```

```python
#7Number of matches won by each team
team_wins = matches['winner'].value_counts()

print(team_wins)

team_wins.plot(kind='bar', color="#CCABD8")
plt.title("Matches Won by Each Team")
plt.show()

#Insight: The max no. of matches is won by Mumbai Indians(109)
         # The least no. of matches is won by Rising Pune Supergiants(5)
```

```python
#8 Number of matches held every season
season_matches = matches['season'].value_counts().sort_index()

print(season_matches)

season_matches.plot(kind='line', marker='o', color="yellow")
plt.title("Matches Played Each Season")
plt.xlabel("Season")
plt.ylabel("Matches")
plt.show()

#Insight:The number of matches increased in later IPL seasons due to the addition of more teams and tournament expansion. The max no. of matches held in a season is 76 which was in 2013.
         # The least no. of matches held in a season is57 in 2009.
```

```python
#9Top 10 players based on Man of the Match awards
top_players = matches['player_of_match'].value_counts().head(10)

print(top_players)

top_players.plot(kind='bar',color='pink')
plt.title("Top 10 Players with Most Man of the Match Awards")
plt.show()
```

```python
#9Top10players based on Man of the Match awards
top_players = matches['player_of_match'].value_counts().head(10)

print(top_players)

top_players.plot(kind='bar',color="#FFA3A6")
plt.title("Top 10 Players with Most Man of the Match Awards")
plt.show()

#Insight: The most awards for man of the match goes to  CH Gayle.
```

```python
#10Does winning the toss mean winning the match?
toss_match = matches[matches['toss_winner'] == matches['winner']]

percentage = (len(toss_match) / len(matches)) * 100

print("Percentage of matches where toss winner also won:", percentage)
#Insight:Winning the toss does not guarantee winning the match, but it can provide a slight advantage depending on pitch and match conditions.
```

```python
#11 Decision taken after winning tos
toss_decision = matches['toss_decision'].value_counts()

print(toss_decision)

toss_decision.plot(kind='pie', autopct='%1.1f%%')
plt.title("Toss Decision Distribution")
plt.show()

#Insights:Teams often choose to field first after winning the toss because chasing a target can be easier in many IPL matches.
```

```python
#12Matches where D/L applied every season
dl_season = matches[matches['dl_applied'] == 1].groupby('season').size()

print(dl_season)

dl_season.plot(kind='bar', color='#6E89A2')
plt.title("D/L Method Applied per Season")
plt.show()
#Insight:The D/L method is used only in some seasons when matches are affected by rain.
```

```python
#13Different results for matches
results = matches['result'].value_counts()

print(results)

results.plot(kind='bar',color="#D6A8C4")
plt.title("Types of Match Results")
plt.show()

#Insight:Most IPL matches end in normal results, while ties and no-result matches are very rare# .# The results were like this
# Normal result: 743 times
# Tie result : 9 times
# No result : 4 times
```

```python
#14How many times did a team win the toss
toss_wins = matches['toss_winner'].value_counts()

print(toss_wins)

toss_wins.plot(kind='bar',color='#8EB69B')
plt.title("Toss Wins by Teams")
plt.show()
# shows how often each team had the advantage of deciding whether to bat or field first.
#Insight :# Max no. of times Mumbai Indians won the toss which is for 98 times
# Min no.of times Rising Pune Supergiant won the toss which is for 6 times only.
```

```python
#15Best venue for defending total
defend = matches[matches['win_by_runs'] > 0]
best_defend_venue = defend['venue'].value_counts().head(10)

print(best_defend_venue)

#Insight:Some stadiums favor bowlers or have pitch conditions where defending a score becomes easier.
```

```python
#16Best defending teams
best_defending_team = defend['winner'].value_counts().head(10)

print(best_defending_team)
#Insight:These teams have strong bowling attacks and strategies that help them successfully defend targets.
```

```python
#17Matches played in different stadiums
stadium_matches = matches['venue'].value_counts()

print(stadium_matches)

stadium_matches.plot(kind='bar', figsize=(10,5),color='#C4B6AB')
plt.title("Matches in Different Stadiums")
plt.show()

#Insight:Popular cricket stadiums host more matches because of their large capacity and better facilities.
```
